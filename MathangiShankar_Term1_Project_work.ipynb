### 4.10 Calculate max and min stats to know the Pokemon which highest and lowest stats

```python
stats=['Total','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
```

```python
def maxStat(Poke,column):
    statement=''
    for col in column:
        stat = Poke[col].max()
        #print(stat)
        #print(type(Poke[Poke[col]==Poke[col].max()]['Name'].values[0])) # Names linked with maximum stat
        name = Poke[Poke[col]==Poke[col].max()]['Name'].values[0]
        gen =  Poke[Poke[col]==Poke[col].max()]['Generation'].values[0] # Gen linked with maximum stat
        statement +=name+ ' of Generation ' + str(gen)+' has the best '+col+' stat of '+str(stat)+'\n'
    return statement
print(maxStat(Poke,stats))

def minStat(Poke,column):
    statement=''
    for col in column:
        stat = Poke[col].min()
        #print(stat)
        #print(type(Poke[Poke[col]==Poke[col].max()]['Name'].values[0])) # Names linked with maximum stat
        name = Poke[Poke[col]==Poke[col].min()]['Name'].values[0]
        gen =  Poke[Poke[col]==Poke[col].min()]['Generation'].values[0] # Gen linked with maximum stat
        statement +=name+ ' of Generation ' + str(gen)+' has the worst '+col+' stat of '+str(stat)+'\n'
    return statement
print(minStat(Poke,stats))
```
